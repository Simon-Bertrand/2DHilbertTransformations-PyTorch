```python
import torch,  numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = torch.Tensor(np.array(Image.open("example_img.gif")))
plt.imshow(img)
```

# Basic Hilbert Transform

```python
print("Basic Hilbert Transform")
plt.imshow(torch.abs(torch_hilbert.Basic.computeHT(img)))
```

```python
print("Basic Analytic Signal")
plt.imshow(torch.abs(torch_hilbert.Basic.computeAS(img)))
```

# Directional Hilbert Transform

```python
print("Directional Hilbert Transform, Axis=0")
plt.imshow(torch.abs(torch_hilbert.Directional.computeHT(img, axis=0)))
plt.show()

print("Directional Hilbert Transform, Axis=1")
plt.imshow(torch.abs(torch_hilbert.Directional.computeHT(img, axis=1)))
plt.show()
```

```python
print("Directional Analytic Signal, Axis=0")
plt.imshow(torch.abs(torch_hilbert.Directional.computeAS(img, axis=0)))
plt.show()

print("Directional Analytic Signal, Axis=1")
plt.imshow(torch.abs(torch_hilbert.Directional.computeAS(img, axis=1)))
plt.show()
```

# Single Orthant

```python
print("Single Orthant Hilbert Transform")
plt.imshow(torch.abs(torch_hilbert.SingleOrthant.computeHT(img)))
```

```python
print("Single Orthant Analytic Signal")
plt.imshow(torch.abs(torch_hilbert.SingleOrthant.computeAS(img)))
```

# Combined Hilbert Transform

```python
print("Combined Hilbert Transform, Axis=0")
plt.imshow(torch.abs(torch_hilbert.Combined.computeHT(img, axis=0)))
plt.show()

print("Combined Hilbert Transform, Axis=1")
plt.imshow(torch.abs(torch_hilbert.Combined.computeHT(img, axis=1)))
plt.show()
```

```python
print("Combined Analytic Signal, Axis=0")
plt.imshow(torch.abs(torch_hilbert.Combined.computeAS(img, axis=0)))
plt.show()

print("Combined Analytic Signal, Axis=1")
plt.imshow(torch.abs(torch_hilbert.Combined.computeAS(img, axis=1)))
plt.show()
```
